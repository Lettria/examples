#### Extract text chunks

```python
import requests

api_host = 'https://api.lettria.com'
api_route = api_host + '/parse'
jwt_token_file = 'jwt_token.txt'
jwt_token = open(jwt_token_file, 'r').read().strip()

text = """
Francois est un jeune homme de 25 ans.
Il est né à Paris.
"""

response = requests.post(
    api_route,
    json={
        "text": text,
    },
    headers={"Authorization":"LettriaProKey " + jwt_token}
)
response = response.json()

text_chunks = [chunk['content'] for chunk in response['chunks'] if chunk['type'] == 'text']
print(text_chunks)
```

#### Text to graph using Lettria structuration endpoint

```python
api_route = api_host + '/structuration'

response = requests.post(
    api_route,
    json={
        "documents": text_chunks,
        "options":{
            "rdf_export": True
        }
    },
    headers={"Authorization":"LettriaProKey " + jwt_token}
).json()

rdf_string = response['rdf']
print(rdf_string)
```
